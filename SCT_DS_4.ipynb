```python
import pandas as pd

df = pd.read_csv('US_Accidents_March23.csv', nrows=100000)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df['Severity'].value_counts().plot(kind='bar')

plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Start_Lng'], df['Start_Lat'], s=1, alpha=0.3)

plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

df['Severity'].value_counts().sort_index().plot(kind='bar')

plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
df['Weather_Condition'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Weather Conditions')
plt.xlabel('Weather')
plt.ylabel('Accidents')
plt.show()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['Hour'] = df['Start_Time'].dt.hour

df['Hour'].value_counts().sort_index().plot(kind='line')

plt.title('Accidents by Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['State'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 States by Accidents')
plt.xlabel('State')
plt.ylabel('Accidents')
plt.show()
```

```python
road_features = ['Traffic_Signal','Junction','Crossing','Stop']

for col in road_features:
    print(col)
    print(df[col].value_counts())
    print()
```
